```python
import os
import torch
import torch.utils.data as torchdata
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
import tqdm
import torch.optim as optim
import torch.backends.cudnn as cudnn
cudnn.benchmark = True
import argparse
from torch.autograd import Variable
# from tensorboard_logger import configure, log_value
from torch.distributions import Bernoulli

from utils import utils, utils_detector
from constants import base_dir_metric_cd, base_dir_metric_fd
from constants import num_actions

import warnings
warnings.simplefilter("ignore")

import matplotlib.pyplot as plt
from PIL import Image
import cv2
import torchvision.transforms as transforms

# GPU Device
gpu_id = '0'
os.environ['CUDA_VISIBLE_DEVICES'] = str(gpu_id)
use_cuda = torch.cuda.is_available()
print("GPU device " , use_cuda)
```

```python
num_actions = 4
pt_model = 'best_model_256_A2C.pth'
img_name = 'ship02017071802012016.jpg'
```

# Image Load

```python
# load image sample
# img_dir = './Demo_sample/' + img_name
img_dir = './Demo_sample/' + img_name
img = Image.open(img_dir)
plt.figure(figsize=(10,10))
plt.imshow(img)
img_size = 256
mean = [0.485, 0.456, 0.406]
std = [0.229, 0.224, 0.225]
transform_test = transforms.Compose([
        transforms.Scale(img_size),
        transforms.CenterCrop(img_size),
        transforms.ToTensor(),
        transforms.Normalize(mean, std)
    ])
```

# Pre-trained model load

```python
trainset, testset = utils.get_dataset(128, './data/256/')
agent = utils.get_model(num_actions)
load_model = torch.load(pt_model)
agent.load_state_dict(load_model['agent'])
```

```python
inputs = transform_test(img).cuda().unsqueeze(0)
agent = agent.cuda().eval()
```

```python
# inputs = i[0].cuda()
```

```python
# Actions by the Policy Network
probs = F.sigmoid(agent(inputs))

# Sample the policy from the agents output
policy = probs.data.clone()
policy[policy<0.5] = 0.0
policy[policy>=0.5] = 1.0
policy = Variable(policy)
```

```python
print(probs)
print(policy)
```

# Choosen Policy Visualization

```python
imgs = []
for i in range(4):
    imgs.append(img_name.split('.')[0]+'_'+str(i)+'.jpg')

policies = policy.cpu().numpy()[0]


def img_path(img_name, i):    
    if policies[i] == 1:
        return './Demo_sample/160/' + img_name
    else:
        return './Demo_sample/32/' + img_name
    
img1 = Image.open(img_path(imgs[0], 0))
img2 = Image.open(img_path(imgs[1], 1))
img3 = Image.open(img_path(imgs[2], 2))
img4 = Image.open(img_path(imgs[3], 3))
img1_size = img1.size
img2_size = img2.size
img3_size = img3.size
img4_size = img4.size

fig = plt.figure()
fig, ax_lst = plt.subplots(2, 2, figsize=(8,8))
for i in range(2):
    for j in range(2):
        ax_lst[i][j].set_xticks([])
        ax_lst[i][j].set_yticks([])
ax_lst[0][0].set_title('Policy 1: {} Size'.format(img1_size))
ax_lst[0][0].imshow(img1)
ax_lst[0][1].set_title('Policy 2: {} Size'.format(img2_size))
ax_lst[0][1].imshow(img2)
ax_lst[1][0].set_title('Policy 3: {} Size'.format(img3_size))
ax_lst[1][0].imshow(img3)
ax_lst[1][1].set_title('Policy 4: {} Size'.format(img4_size))
ax_lst[1][1].imshow(img4)
plt.show()
```

# Bounding Box from YOLO

```python
imgs = []
for i in range(2):
    for j in range(2):
        imgs.append(img_name.split('.')[0]+'_'+str(i)+'_'+str(j)+'.npy')


def box_path(img_name, i):    
    if policies[i] == 1:
        return 'data/256/base_dir_detections_fd/' + img_name
    else:
        return 'data/256/base_dir_detections_cd/' + img_name

target1 = np.load(box_path(imgs[0], 0))
target2 = np.load(box_path(imgs[1], 1))
target3 = np.load(box_path(imgs[2], 2))
target4 = np.load(box_path(imgs[3], 3))
bbox1_1, bbox1_2, bbox2_1, bbox2_1, bbox3_1, bbox3_2, bbox4_1, bbox4_2 = [], [], [], [], [], [], [], []
```

```python
for t1 in target1:
    bbox1_1, bbox1_2 = (int(t1[0]), int(t1[1])), (int(t1[2]), int(t1[3]))
    img1 = cv2.rectangle(np.array(img1), bbox1_1, bbox1_2, (255,0,0), 1)
for t2 in target2:
    bbox2_1, bbox2_2 = (int(t2[0]), int(t2[1])), (int(t2[2]), int(t2[3]))
    img2 = cv2.rectangle(np.array(img2), bbox2_1, bbox2_2, (255,0,0), 1)
for t3 in target3:
    bbox3_1, bbox3_2 = (int(t3[0]), int(t3[1])), (int(t3[2]), int(t3[3]))
    img3 = cv2.rectangle(np.array(img3), bbox3_1, bbox3_2, (255,0,0), 1)
for t4 in target4:
    bbox4_1, bbox4_2 = (int(t4[0]), int(t4[1])), (int(t4[2]), int(t4[3]))
    img4 = cv2.rectangle(np.array(img4), bbox4_1, bbox4_2, (255,0,0), 1)
```

```python
fig = plt.figure()
fig, ax_lst = plt.subplots(2, 2, figsize=(8,8))
for i in range(2):
    for j in range(2):
        ax_lst[i][j].set_xticks([])
        ax_lst[i][j].set_yticks([])
ax_lst[0][0].set_title('Policy 1: {} Size'.format(img1_size))
ax_lst[0][0].imshow(img1)
ax_lst[0][1].set_title('Policy 2: {} Size'.format(img2_size))
ax_lst[0][1].imshow(img2)
ax_lst[1][0].set_title('Policy 3: {} Size'.format(img3_size))
ax_lst[1][0].imshow(img3)
ax_lst[1][1].set_title('Policy 4: {} Size'.format(img4_size))
ax_lst[1][1].imshow(img4)
plt.show()
```
